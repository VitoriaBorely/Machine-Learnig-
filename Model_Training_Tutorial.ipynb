##**Challenge 01**
Encode the categorical variables from the following dataset.


```python
import pandas as pd 

df = pd.DataFrame({'Car':['Honda', 'Toyota', 'Toyota', 'Peugeot', 'Ford'],
                   'Year':['2000', '2010', '2009', '2008', '2013'], 
                   'Price':[17000, 23000, 22000, 20000, 24000]})
df
```

```python
from sklearn.model_selection import train_test_split 

from sklearn.preprocessing import LabelEncoder 

encoder = LabelEncoder()
encoder = encoder.fit(df['Car'])
encoded = encoder.transform(df['Car'])

df['Car'] = encoded 

df
```
