#**Challenge 6**
# Change the column 'Sex', so that the observations with the value 'male' are replaced by the value '1' and observations with the value 'female' are replaced by the value '0'

```python
import pandas as pd 
url = "https://raw.githubusercontent.com/pmarcelino/datasets/master/titanic.csv"
df = pd.read_csv(url)
df
```

```python
df.loc['Sex']
```

```python
df.loc['Sex' , 'male'] = 1 
df.loc['Sex', 'female'] = 0
df
```
